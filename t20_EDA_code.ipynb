BOWLING SUMMARY

```python
with open('t20_json_files/t20_wc_bowling_summary.json') as f2:
    data2 = json.load(f2)
pd_bowling = pd.DataFrame(data2[0]['bowlingSummary'])
pd_bowling.head()
```

```python
pd_bowling['overs']=pd_bowling['overs'].astype('int')
pd_bowling['TotalBalls']=pd_bowling['overs']*6
pd_bowling.head()
```

```python
pd_bowling['wickets']=pd_bowling['wickets'].astype('int')
pd_bowling['BStrickeRate']=pd_bowling['TotalBalls']/pd_bowling['wickets']
pd_bowling.head()
```

```python
pd_bowling2=pd_bowling.where(pd_bowling['BStrickeRate']<20)
pd_bowling2.fillna(0,inplace=True)
pd_bowling2.head()
```

```python
pd_bowling2['bowlerName']=pd_bowling2['bowlerName'].astype('str')
x = pd_bowling2['bowlerName']
y = pd_bowling2['BStrickeRate']
plt.bar(x,y)
ax = plt.gca()
ax.tick_params(axis='x', labelrotation = 45)
plt.show()
```

```python
x = pd_bowling2['bowlerName']
y = pd_bowling2['economy'].astype('float')
plt.bar(x,y)
ax = plt.gca()
ax.tick_params(axis='x', labelrotation = 45)
plt.show()
```

**BATTING SUMMARY**

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f3:
    data3 = json.load(f3)
    
    batting_records = []

    for rec in data3:
        batting_records.extend(rec['battingSummary'])
        
batting_records
pd_batting = pd.DataFrame(batting_records)
pd_batting.head()
```

```python
pd_batting["out/not_out"] = pd_batting.dismissal.apply(lambda x : "out" if len(x)>0 else "not-out")
pd_batting.head(30)
```

```python
pd_batting.drop(columns = ['dismissal'], inplace = True)
pd_batting['batsmanName'] = pd_batting['batsmanName'].apply(lambda x : x.replace('',''))
pd_batting['batsmanName'] = pd_batting['batsmanName'].apply(lambda x : x.replace('/',''))
pd_batting.head()
```

```python
match_ids_dict = {}

for index, row in pd_match.iterrows():
    key1 = row['team1']+ ' Vs '+ row['team2']
    key2 = row['team2']+ ' Vs '+ row['team1']
    
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
    
match_ids_dict
```

```python
pd_batting["match_id"] = pd_batting["match"].map(match_ids_dict)

pd_batting.head()
```

```python
pd_batting['battingPos']=pd_batting['battingPos'].astype('int')
type(pd_batting['battingPos'])
```

```python
pd_batting['4s']=pd_batting['4s'].astype('int')
pd_batting['6s']=pd_batting['6s'].astype('int')
pd_batting['runs']=pd_batting['runs'].astype('int')
```

```python
for i in range(0,len(pd_batting)):
    if(pd_batting.loc[i, "SR"]=='-'):
        pd_batting.loc[i, "SR"]='0'
        
pd_batting.head()
```
